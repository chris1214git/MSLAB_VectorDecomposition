### PubMed

```python
import os

dir_path = '../data/PubMed/docs'

texts = []
for file in os.listdir(dir_path):
    file = os.path.join(dir_path, file)
    with open(file, 'r') as f:
        try:
            text = ''.join(f.readlines())
            
        except:
            print('error', file)
            continue
    text = text.replace('\n', ' ')
    texts.append(text)
    
```

```python
with open('../data/PubMed.txt', 'w') as f:
    for text in texts:
        print(text, file = f)
```
